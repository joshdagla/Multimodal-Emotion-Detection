```python
import torch
import ultralytics
import os
```

```python
# Path to dataset and data.yaml
DATA_YAML = r"C:/Users/joshd/Documents/BTech Project Sem-7/btech_project/facial_emotion_detection/facial_emotion_detection_9400/Human face emotions.v30--prueba-2-tfg-implementacion-amb82mini-yolv7-tiny.yolov11/data.yaml"

assert os.path.exists(DATA_YAML), f"{DATA_YAML} not found!"
```

```python
from ultralytics import YOLO

# Load YOLOv11 base model (n=Nano, m=Medium, l=Large)
model = YOLO("yolo11n.pt")  # For speed. For higher accuracy, try yolo11m.pt or yolo11l.pt

# Train the model
model.train(
    data=DATA_YAML,
    epochs=80,
    imgsz=640,
    batch=8,
    project="runs/train",          # Training results will be stored here
    name="facial_emotion_detection_model", # Custom folder name
    device=0,           # GPU 0 (use "cpu" if no CUDA)
    amp=True
)
```

```python
# Validate the trained model
metrics = model.val(data=DATA_YAML)
print("Validation Metrics:", metrics)

# Run inference on test set
results = model.predict(
    source="C:/Users/joshd/Documents/BTech Project Sem-7/btech_project/facial_emotion_detection/facial_emotion_detection_9400/Human face emotions.v30--prueba-2-tfg-implementacion-amb82mini-yolv7-tiny.yolov11/test/images",
    save=True,
    conf=0.25  # Confidence threshold (adjust if needed)
)

print("Results saved. Check runs/predict/emotion_detection_yolo folder")
print("mAP@0.5:", metrics.box.map)   
print("mAP@0.5 (same as map50):", metrics.box.map50)
print("mAP@0.75:", metrics.box.map75)
print("Per-class mAP:", metrics.box.maps)
```
